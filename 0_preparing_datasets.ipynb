```python
import pandas as pd

# Membaca data dari file CSV
df = pd.read_csv('datasets/harakat/Labelling AL-Fatihah.csv', delimiter=';')

# Mengambil data dari kolom 'audio' dan menyimpannya sebagai list
audio_list = df['audio'].tolist()

# Menampilkan hasil
print(audio_list)
```

```python
import os
import shutil

# Path ke folder yang berisi file audio
source_folder = 'datasets/harakat'
destination_folder = 'datasets/harakat_raw'

# Pastikan folder tujuan ada
os.makedirs(destination_folder, exist_ok=True)

# # List nama file audio yang ingin dicari
# audio_list = [
#     "18395a1a-zaki-daghistani_1.mp3",
#     "2718d56f-yousuf-alshoaey_58.mp3"
#     # Tambahkan nama file lainnya
# ]

# Loop melalui setiap nama file dalam audio_list
for audio_filename in audio_list:
    # Path lengkap file audio di folder asal
    source_path = os.path.join(source_folder, audio_filename)

    # Periksa apakah file ada di folder asal
    if os.path.exists(source_path):
        # Path lengkap untuk memindahkan file ke folder tujuan
        destination_path = os.path.join(destination_folder, audio_filename)

        # Pindahkan file
        shutil.move(source_path, destination_path)
        print(f"{audio_filename} berhasil dipindahkan ke {destination_folder}.")
    else:
        print(f"{audio_filename} tidak ditemukan di {source_folder}.")
```

```python
import os
from pydub import AudioSegment

# Folder asal tempat file MP3 berada
source_folder = 'datasets/harakat_raw'
# Folder tujuan untuk menyimpan file WAV yang telah dikonversi
destination_folder = 'datasets/harakat_raw'

# Pastikan folder tujuan ada
os.makedirs(destination_folder, exist_ok=True)

# Loop melalui setiap file dalam folder asal
for filename in os.listdir(source_folder):
   
    if filename.endswith(".mp3"):
        try:
            # Path lengkap file MP3
            mp3_path = os.path.join(source_folder, filename)
            print(mp3_path)
            # Membaca file MP3
            audio = AudioSegment.from_mp3(mp3_path)

            # Mengubah nama file dari .mp3 menjadi .wav
            wav_filename = os.path.splitext(filename)[0] + '.wav'
            wav_path = os.path.join(destination_folder, wav_filename)

            # Menyimpan file sebagai WAV
            audio.export(wav_path, format="wav")
            print(f"{filename} berhasil dikonversi menjadi {wav_filename} dan disimpan di {destination_folder}.")

            # Menghapus file MP3 setelah konversi
            os.remove(mp3_path)
            print(f"{filename} telah dihapus dari {source_folder}.")
        except Exception as e:
            print(e)
    else:
        print(f"{filename} bukan file MP3, dilewati.")
```

```python
import os
import csv
import json
from pydub import AudioSegment

# Path ke file CSV dan folder tempat audio berada
csv_file_path = "datasets/Labelling AL-Fatihah.csv"
audio_folder = "datasets/harakat_raw"
output_folder = "datasets/harakat_raw/labels"

# Membuat output folder jika belum ada
os.makedirs(output_folder, exist_ok=True)

# Membaca file CSV
with open(csv_file_path, mode='r', newline='') as csvfile:
    reader = csv.DictReader(csvfile,delimiter=';')
    
    for row in reader:
        audio_file = row['audio']
        audio_path = os.path.join(audio_folder, audio_file)
        
        try:
            # Load audio file
            audio = AudioSegment.from_file(audio_path)
        except Exception as e:
            print(e)
            
        # Parse JSON dari kolom 'label'
        labels = json.loads(row['label'])
        
        for i, label_data in enumerate(labels):
            start_time = label_data['start'] * 1000  # Convert to milliseconds
            end_time = label_data['end'] * 1000  # Convert to milliseconds
            label_name = label_data['labels'][0]
            
            # Memotong audio
            segment = audio[start_time:end_time]
            
            # Membuat folder berdasarkan label jika belum ada
            label_folder = os.path.join(output_folder, label_name)
            os.makedirs(label_folder, exist_ok=True)
            
            # Menyimpan file hasil potongan
            output_file_path = os.path.join(label_folder, f"{os.path.splitext(audio_file)[0]}_{i}.wav")
            segment.export(output_file_path, format="wav")

            print(f"Saved {output_file_path}")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Path ke folder dataset
base_folder = 'datasets/harakat_raw'
class_folders = ['Madd Arid Lissukun', 'Madd Lazim Mutsaqal Kilmi', 'Madd Thabii']

# Membuat folder tujuan jika belum ada
for class_name in class_folders:
    os.makedirs(os.path.join(base_folder, 'train', class_name), exist_ok=True)
    os.makedirs(os.path.join(base_folder, 'val', class_name), exist_ok=True)

try:
    # Membagi file dalam setiap folder klasifikasi
    for class_name in class_folders:
        class_folder = os.path.join(base_folder, class_name)
        train_folder = os.path.join(base_folder, 'train', class_name)
        val_folder = os.path.join(base_folder, 'val', class_name)
        
        # Mengambil semua file dari class_folder
        all_files = os.listdir(class_folder)

        # Mengacak urutan file dan membagi dataset
        train_files, val_files = train_test_split(all_files, test_size=0.2, random_state=42)

        # Memindahkan file ke folder train
        for file_name in train_files:
            src_path = os.path.join(class_folder, file_name)
            dst_path = os.path.join(train_folder, file_name)
            shutil.move(src_path, dst_path)

        # Memindahkan file ke folder val
        for file_name in val_files:
            src_path = os.path.join(class_folder, file_name)
            dst_path = os.path.join(val_folder, file_name)
            shutil.move(src_path, dst_path)

        print(f"Folder {class_name} dipindahkan menjadi {len(train_files)} file untuk train dan {len(val_files)} file untuk val.")
except Exception as e:
    print(e)
```

```python
import os

def rename_files_with_folder_name(base_path):
    # Mengiterasi semua folder dalam base_path
    for root, dirs, files in os.walk(base_path):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            
            # Menentukan nama folder (subfolder)
            folder_name = os.path.basename(root)
            
            # Memeriksa apakah nama file mengandung unique code yang perlu diganti
            if '-' in file_name:
                unique_code = file_name.split('-')[0]  # Mengambil unique code dari nama file
                new_file_name = file_name.replace(unique_code, folder_name, 1)  # Mengganti unique code dengan nama folder
                
                # Membuat path baru untuk file
                new_file_path = os.path.join(root, new_file_name)
                
                # Menangani nama file yang sudah ada
                counter = 1
                while os.path.exists(new_file_path):
                    base, ext = os.path.splitext(new_file_name)
                    new_file_name = f"{base}_{counter}{ext}"
                    new_file_path = os.path.join(root, new_file_name)
                    counter += 1
                
                # Mengganti nama file
                os.rename(file_path, new_file_path)
                print(f'File "{file_name}" diubah menjadi "{new_file_name}"')

# Path dasar ke folder dataset Anda
base_path = 'C:\\Users\\dimas\\Documents\\GitHub\\panns_transfer_to_qt\\datasets\\harakat_raw\\val'
rename_files_with_folder_name(base_path)
```
